```python

try:
    # 웹 드라이버 초기화
    driver = webdriver.Chrome()
    
    # 페이지 접속
    driver.get('https://edu.tsherpa.co.kr/LectureInfo/LectureReviews2')

    # 첫 번째 팝업 열기
    try:
        first_popup_link = WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="mainForm"]/table/tbody/tr[2]/td[2]/a'))
        )
        first_popup_link.click()
    except (TimeoutException, NoSuchElementException) as e:
        print("첫 번째 팝업 링크를 클릭하는 동안 오류 발생:", e)
        raise

    # 두 번째 팝업으로 전환
    popup_link = WebDriverWait(driver, 30).until(
        EC.element_to_be_clickable((By.XPATH, '//*[@id="mainForm"]/div[1]/span/a'))
    ).get_attribute('href')

    # requests를 사용하여 팝업 페이지 접속 후 BeautifulSoup로 파싱
    content = requests.get(popup_link).content
    soup = BeautifulSoup(content, 'html.parser')

    # 제목(title) 추출
    title = soup.select_one('div.title').text.strip()
    title = title[len(soup.select_one('div.title span').text):].strip()  # 시간 부분을 제외하고 추출

    # 시간(time) 추출
    time = soup.select_one('div.title span').text.strip()

    # 대상(연수종별) 추출
    people = soup.select_one('td').find_next_sibling('td').text.strip()

    # 연수비용 추출
    pay = soup.select('div.info_type td[colspan="3"]')[2].text.strip()
    pay = pay.replace('\r\n                        정가 : ', '').replace('원\r\n                    ', '')

    # 연수분야 추출
    field = soup.select('td')[1].text.strip()

    # 결과 출력
    print("제목:", title)
    print("시간:", time)
    print("대상(연수종별):", people)
    print("연수비용:", pay)
    print("연수분야:", field)

except WebDriverException as e:
    print("웹드라이버 오류:", e)
```

```python
try:
    # 웹 드라이버 초기화
    driver = webdriver.Chrome()

    # 페이지 접속
    driver.get('https://edu.tsherpa.co.kr/LectureInfo/LectureReviews2')

    # 첫 번째 팝업 열기
    try:
        first_popup_link = WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="mainForm"]/table/tbody/tr[2]/td[2]/a'))
        )
        first_popup_link.click()
        print("첫 번째 팝업 링크 클릭 완료")
    except (TimeoutException, NoSuchElementException) as e:
        print("첫 번째 팝업 링크를 클릭하는 동안 오류 발생:", e)
        raise

    # 두 번째 팝업으로 전환
    try:
        WebDriverWait(driver, 30).until(
            EC.number_of_windows_to_be(2)
        )
        driver.switch_to.window(driver.window_handles[1])
        print("두 번째 팝업 창으로 전환 완료")
    except TimeoutException as e:
        print("두 번째 팝업으로 전환하는 동안 오류 발생:", e)
        raise

    # 팝업 내 링크 클릭하여 새 창으로 이동
    try:
        link_in_popup = WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="mainForm"]/div[1]/span/a'))
        )
        link_in_popup.click()
        print("팝업 내 링크 클릭 완료")
    except (TimeoutException, NoSuchElementException) as e:
        print("팝업 내 링크 클릭하는 동안 오류 발생:", e)
        raise

    # 새로 열린 창으로 전환
    try:
        WebDriverWait(driver, 60).until(EC.number_of_windows_to_be(3))  # 세 번째 창이 열릴 때까지 기다립니다.
        print("세 번째 창이 열림 확인. 현재 창 핸들:", driver.window_handles)
        third_window_handle = driver.window_handles[2]
        driver.switch_to.window(third_window_handle)
        print("세 번째 창으로 전환 완료:", third_window_handle)
    except TimeoutException as e:
        print("TimeoutException 발생:", e)
        raise
    except NoSuchWindowException as e:
        print("NoSuchWindowException 발생:", e)
        raise
    except WebDriverException as e:
        print("WebDriverException 발생:", e)
        raise

    # 새 창에서 추가 작업 수행
    print("새 창에서 작업 수행 중...")

    # 예를 들어, 페이지의 타이틀 출력
    print("새 창 타이틀:", driver.title)

except Exception as e:
    print("오류 발생:", e)
finally:
    driver.quit()
```
